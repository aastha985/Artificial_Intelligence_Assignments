# Importing Modules

```python
import pandas as pd
import numpy as np
from collections import defaultdict
```

# Importing the dataset


```python
df=pd.read_csv('roaddistance.csv')
```

```python
df.columns = df.iloc[0]
df = df.iloc[1:,1:-1]
df = df.replace('-',0)
cities = df['Distance in Kilometres']
```

```python
df.head()
```

# Assigning Heuristic

Heuristic is assigned as the shortest distance between 2 pairs of cities using floyd warshall algo

```python
#floyd warshall algorithm

distance = defaultdict(dict)

for city1 in cities:
  for city2 in cities:
    if city1==city2:
      distance[city1][city1] = 0
    else:
      distance[city1][city2] = 1e9
      distance[city2][city1] = 1e9

for city in df.columns.values[1:]:
  for i in range(1,len(df[city])+1):
    city1 = city
    city2 = cities[i]
    weight = int(df[city][i])

    distance[city1][city2] = weight
    distance[city2][city1] = weight

for k in cities:
  for i in cities:
    for j in cities:
      if distance[i][k]!=1e9 and distance[k][j]!=1e9 and (distance[i][j] > distance[i][k] + distance[k][j]):
        distance[i][j] = distance[i][k] + distance[k][j]
```

```python
heuristics = []

for city1 in cities:
  for city2 in cities:
    if city1==city2:
      continue
    heuristics.append({
        "City 1": city2,
        "City 2": city1,
        "Heuristic": distance[city1][city2]})

df2 = pd.DataFrame(heuristics)
df2.to_csv("Heuristic.csv",index=False)
```

```python
distances = []

for city in df.columns.values[1:]:
  for i in range(1,len(df[city])+1):
    city1 = city
    city2 = cities[i]
    dist = int(df[city][i])

    distances.append({
        "City 1": city2,
        "City 2": city1,
        "Distance" : dist
    })

df3 = pd.DataFrame(distances)
df3.to_csv("Distance.csv",index=False)
```
